```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
# Parameters
lambda_ = 5  # average rate of events (e.g., 5 events per hour)


# Parameters
scale = 1 / lambda_  # 1 / rate of occurrence

# Range for plotting
x_values = np.linspace(0, 10, 1000)

# PDF and CDF
pdf_values = expon.pdf(x_values, scale=scale)
cdf_values = expon.cdf(x_values, scale=scale)

# Plotting
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x_values, pdf_values, color='blue', label='PDF')
plt.title("Exponential Distribution - PDF")
plt.xlabel("Time")
plt.ylabel("Density")
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_values, cdf_values, color='green', label='CDF')
plt.title("Exponential Distribution - CDF")
plt.xlabel("Time")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
```
