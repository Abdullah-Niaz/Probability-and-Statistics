```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for normal distribution
mean = 0    # mean of the distribution
std_dev = 1 # standard deviation of the distribution

# Range of values for x-axis
x_values = np.linspace(-4, 4, 1000)

# Compute PDF and CDF
pdf_values = norm.pdf(x_values, mean, std_dev)
cdf_values = norm.cdf(x_values, mean, std_dev)

# Plotting
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x_values, pdf_values, color='blue', label='PDF')
plt.title("Normal Distribution - PDF")
plt.xlabel("X-axis")
plt.ylabel("Density")
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_values, cdf_values, color='green', label='CDF')
plt.title("Normal Distribution - CDF")
plt.xlabel("X-axis")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
```
