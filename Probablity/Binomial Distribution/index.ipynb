```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for binomial distribution
n = 10    # number of trials (e.g., coin tosses)
p = 0.5   # probability of success (e.g., probability of getting heads)

# Possible outcomes
x_values = np.arange(0, n+1)

# Compute PMF and CDF
pmf_values = binom.pmf(x_values, n, p)
cdf_values = binom.cdf(x_values, n, p)

# Plotting
plt.figure(figsize=(12, 6))

# Plot PMF
plt.subplot(1, 2, 1)
plt.stem(x_values, pmf_values, basefmt=" ", linefmt='blue', markerfmt='bo', label='PMF')
plt.title("Binomial Distribution - PMF")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x_values, cdf_values, color='green', where='post', label='CDF')
plt.title("Binomial Distribution - CDF")
plt.xlabel("Number of Successes")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
```
